```python
from pathlib import Path
import pickle
import pydicom
import numpy as np
```

```python
def get_case_p_slice_locations(files_list, case_name):
    slice_locations = list()
    for file in files_list:
        filename = file['phaseP'].split("\\")[-1]
        dcm = pydicom.read_file(f"./{case_name}/P/{filename}")
        slice_locations.append(float(dcm.SliceLocation))
    return slice_locations


def get_case_p_pixel_spacing(files_list, case_name):
    pixel_spacings = list()
    file = files_list[0]
    filename = file['phaseP'].split("\\")[-1]
    dcm = pydicom.read_file(f"./{case_name}/P/{filename}")
    return np.array(dcm.PixelSpacing, dtype='float')
```

```python
root_pth = Path.cwd()
```

```python
p_pixel_spacing = dict()
for pth in root_pth.glob("file_lists/*.pkl"):
    name = pth.name.replace("-files.pkl", "")
    print(name)
    with open(pth, "rb") as f:
        files_list = pickle.load(f)

    pixel_spacings = get_case_p_pixel_spacing(files_list, name)
    p_pixel_spacing[name] = pixel_spacings
```

```python
with open("p_pixel_spacings.pkl", "wb") as f:
    pickle.dump(p_pixel_spacing, f)
```
