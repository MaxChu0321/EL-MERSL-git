```python
import os
import shutil

# 定義存放資料的路徑和目標資料夾的路徑
data_path = '/Users/mac/Library/CloudStorage/GoogleDrive-maxchu.bt11@nycu.edu.tw/共用雲端硬碟/TVGH-Group/手術復發/TGcode/ISD_Output0902/check_img_補充/HCCA/'  # 資料所在的路徑
target_folder_path = '/Users/mac/Library/CloudStorage/GoogleDrive-maxchu.bt11@nycu.edu.tw/共用雲端硬碟/TVGH-Group/手術復發/TGcode/Output4/check_img/HCCA'  # 目標資料夾的路徑

# 獲取目標資料夾中的資料夾名稱
target_folders = os.listdir(target_folder_path)

# 遍歷資料路徑下的所有檔案
for filename in os.listdir(data_path):
    if filename.endswith('.mat') and filename.startswith('ISD-'):
        # 提取檔案名稱的前半部分
        parts = filename.split('_')
        if len(parts) >= 2:
            prefix = parts[0]
            
            # 檢查是否存在對應的資料夾
            if prefix in target_folders:
                # 建立目標路徑
                target_path = os.path.join(target_folder_path, prefix)
                
                
                # 複製檔案到目標路徑
                shutil.copy(os.path.join(data_path, filename), target_path)
                print(f'已將檔案 {filename} 複製到 {target_path}')
            else:
                print(f'找不到對應的目標資料夾 for 檔案 {filename}')

print('處理完成')
```

```python
import scipy.io
import pickle
import numpy as np
import os
import glob

# 指定根文件夹路径
root_folder = 'D:/共用雲端硬碟/TVGH-Group/手術復發/TGcode/Output4/check_img/'
def process_subfolders(folder_path):
    # 使用glob查找当前文件夹下的所有以_pixelspacing.mat结尾的文件
    mat_files = glob.glob(os.path.join(folder_path, '*_pixelspacing.mat'))
    
    # 遍历匹配的.mat文件
    for mat_file_path in mat_files:
        try:
            # 使用scipy.io.loadmat()方法读取.mat文件
            data = scipy.io.loadmat(mat_file_path)
            
            # 从读取到的数据中提取pixel_spacing
            pixel_spacing = data['pixel_spacing']
            
            # 创建一个1x2的矩阵，将pixel_spacing中的数值填入
            new_data = np.array([pixel_spacing[0, 0], pixel_spacing[1, 0]])
            
            # 創建新的.pkl檔案名稱，保持与原始.mat文件相同的名称
            base_filename = os.path.splitext(os.path.basename(mat_file_path))[0]
            pkl_file_path = os.path.join(folder_path, f'{base_filename}.pkl')
            # dir_path = os.path.dirname(mat_file_path)
    
            # # 創建新的.pkl檔案路徑，保持與原始.mat檔案相同的目錄
            # pkl_file_path = os.path.join(dir_path, f'{base_filename}.pkl')
            # print(pkl_file_path)
            # 使用pickle.dump()方法将新的数据保存为.pkl文件
            with open(pkl_file_path, 'wb') as pkl_file:
                pickle.dump(new_data, pkl_file)
            
            print(f'.pkl文件已成功儲存：{pkl_file_path}')
            
        except Exception as e:
            print(f'处理文件{mat_file_path}时发生错误：{str(e)}')
    
    # 递归处理当前文件夹的子文件夹
    subfolders = [subfolder for subfolder in os.listdir(folder_path) if os.path.isdir(os.path.join(folder_path, subfolder))]
    for subfolder in subfolders:
        subfolder_path = os.path.join(folder_path, subfolder)
        process_subfolders(subfolder_path)

# 列出要处理的子文件夹名称，例如'HCCA'，并调用处理函数
subfolder_name = ['HCCA', 'HCCN', 'HCCP']
for subfolder in subfolder_name:
    subfolder_path = os.path.join(root_folder, subfolder)
    process_subfolders(subfolder_path)
```

```python
import scipy.io
import pickle
import numpy as np
import os

# 指定.mat檔案的路徑
mat_file_path = './ISD-0003_slicelocations.mat'

try:
    # 使用scipy.io.loadmat()方法讀取.mat檔案
    data = scipy.io.loadmat(mat_file_path)
    
    # 從讀取到的資料中提取slice_locations
    slice_locations = data['slice_locations']
    
    # 使用np.ravel()方法將多維數組轉換為一維
    new_data = np.ravel(slice_locations)
    
    # 只儲存一半的數據，使用切片操作
    half_data = new_data[:len(new_data)//2]
    
    # 創建新的.pkl檔案名稱，保持與原始.mat檔案相同的名稱
    base_filename = os.path.splitext(os.path.basename(mat_file_path))[0]
    pkl_file_path = f'{base_filename}_half.pkl'
    
    # 儲存一半的數據到.pkl檔案
    with open(pkl_file_path, 'wb') as pkl_file:
        pickle.dump(half_data, pkl_file)
    
    print(f'一半的.pkl檔案已成功儲存：{pkl_file_path}')
    
except FileNotFoundError:
    print(f'找不到檔案：{mat_file_path}')
except Exception as e:
    print(f'發生錯誤：{str(e)}')
```
