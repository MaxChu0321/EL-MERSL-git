```python
import os
import numpy as np
from PIL import Image

# # HCCA
# hcca_folder_path = '/Users/mac/Library/CloudStorage/GoogleDrive-maxchu.bt11@nycu.edu.tw//TVGH-Group//TGcode/Output4/check_img/HCCN/'  # HCCA
# HCCA
hcca_folder_path = 'D://TVGH-Group//TGcode/Output4/check_img/HCCP/' 
# HCCA
if not os.path.exists(hcca_folder_path):
    print(f"HCCN: {hcca_folder_path}")
else:
    # HCCA
    for isd_folder_name in os.listdir(hcca_folder_path):
        isd_folder_path = os.path.join(hcca_folder_path, isd_folder_name)
        
        # "ISD"
        if os.path.isdir(isd_folder_path) and isd_folder_name.startswith('ISD'):
            # 
            image_files = [f for f in os.listdir(isd_folder_path) if f.endswith('.jpg') or f.endswith('.png')]  # 
            
            if len(image_files) == 0:
                print(f"{isd_folder_name} ")
            else:
                # NumPy
                images = []
                for image_file in image_files:
                    image_path = os.path.join(isd_folder_path, image_file)
                    img = Image.open(image_path)
                    images.append(np.array(img))
                
                # NumPyimgs.npy
                images_array = np.array(images)
                output_path = os.path.join(isd_folder_path, 'imgs.npy')
                np.save(output_path, images_array)
                print(f"{len(images)} {output_path}")
```
