```python
import os
import numpy as np
from PIL import Image

# 指定圖片所在的資料夾路徑
folder_path = '/Users/mac/Library/CloudStorage/GoogleDrive-maxchu.bt11@nycu.edu.tw/共用雲端硬碟/TVGH-Group/手術復發/TGcode/Output4/check_img/HCCA/ISD-0006'  # 請替換為你的實際資料夾路徑

# 檢查資料夾是否存在
if not os.path.exists(folder_path):
    print(f"找不到資料夾: {folder_path}")
else:
    # 讀取資料夾中的所有圖片文件
    image_files = [f for f in os.listdir(folder_path) if f.endswith('.jpg') or f.endswith('.png')]  # 根據你的圖片格式進行調整
    
    if len(image_files) == 0:
        print(f"資料夾中找不到圖片文件: {folder_path}")
    else:
        # 讀取圖片並將它們存儲為一個NumPy陣列
        images = []
        for image_file in image_files:
            image_path = os.path.join(folder_path, image_file)
            img = Image.open(image_path)
            images.append(np.array(img))
        
        # 轉換為NumPy陣列並儲存為imgs.npy檔案
        images_array = np.array(images)
        # np.save(folder_path+'imgs.npy', images_array)
        print(f"成功將{len(images)}張圖片儲存為imgs.npy")
        output_path = os.path.join(folder_path, 'imgs.npy')
        np.save(output_path, images_array)
```

```python
import os
import numpy as np
from PIL import Image

# # 指定HCCA資料夾路徑
# hcca_folder_path = '/Users/mac/Library/CloudStorage/GoogleDrive-maxchu.bt11@nycu.edu.tw/共用雲端硬碟/TVGH-Group/手術復發/TGcode/Output4/check_img/HCCN/'  # 請替換為你的HCCA資料夾路徑
# 指定HCCA資料夾路徑
hcca_folder_path = 'D:/共用雲端硬碟/TVGH-Group/手術復發/TGcode/Output4/check_img/HCCP/' 
# 檢查HCCA資料夾是否存在
if not os.path.exists(hcca_folder_path):
    print(f"找不到HCCN資料夾: {hcca_folder_path}")
else:
    # 遍歷HCCA資料夾下的所有子資料夾
    for isd_folder_name in os.listdir(hcca_folder_path):
        isd_folder_path = os.path.join(hcca_folder_path, isd_folder_name)
        
        # 檢查子資料夾是否是資料夾且以"ISD"開頭
        if os.path.isdir(isd_folder_path) and isd_folder_name.startswith('ISD'):
            # 讀取子資料夾中的所有圖片文件
            image_files = [f for f in os.listdir(isd_folder_path) if f.endswith('.jpg') or f.endswith('.png')]  # 根據你的圖片格式進行調整
            
            if len(image_files) == 0:
                print(f"{isd_folder_name} 資料夾中找不到圖片文件")
            else:
                # 讀取圖片並將它們存儲為一個NumPy陣列
                images = []
                for image_file in image_files:
                    image_path = os.path.join(isd_folder_path, image_file)
                    img = Image.open(image_path)
                    images.append(np.array(img))
                
                # 轉換為NumPy陣列並儲存為imgs.npy檔案
                images_array = np.array(images)
                output_path = os.path.join(isd_folder_path, 'imgs.npy')
                np.save(output_path, images_array)
                print(f"成功將{len(images)}張圖片儲存為 {output_path}")
```
